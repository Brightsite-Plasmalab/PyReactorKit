```python
from pyreactorkit.models import PlugFlowReactor, HeatedPlugFlowReactor
import numpy as np
import matplotlib.pyplot as plt
import wedme.apply.dev

z = np.linspace(0, 500e-3, 1000)
reactor = PlugFlowReactor(T0=300, pressure=50e2, mass_flow_rate_slm=20, 
                          X0={"H2": 1, "H": 1}, 
                          z=z, dz = np.diff(z)[0])

solarr, _, _ = reactor.simulate()
```

```python
from pyreactorkit.display import plot_species, plot_temperature

plot_species(solarr, xaxis="z")
plot_temperature(solarr, xaxis="z")
```
