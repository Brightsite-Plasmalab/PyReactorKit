```python
import pyreactorkit
import collomech

reac = pyreactorkit.models.Reactor0D(chemistry_set=collomech.CH4_METHANE_GRI30_GRIMECH30)

result = reac.run(
    T=3000,
    fix_T=True,
    p=10e2,
    X0={"H2": 0.5, "O2": 0.5},
    t_max=1e-4
)
```

```python
pyreactorkit.display.plot_species(result)
pyreactorkit.display.plot_production_rates(result)

print("")
```
