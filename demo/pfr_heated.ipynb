```python
from pyreactorkit.models import PlugFlowReactor, HeatedPlugFlowReactor
import numpy as np
import matplotlib.pyplot as plt
import wedme.apply.dev

z = np.linspace(0, 50e-3, 1000)
reactor = HeatedPlugFlowReactor(T0=300, pressure=50e2, mass_flow_rate_slm=10, 
                          X0={"CO2": 1}, 
                          z=z, dz = np.diff(z)[0])

heating_profile = (z < 5e-3) * 1.0
heating_profile *= 2000 / np.sum(heating_profile)

reactor.set_heating_profile(heating_profile=heating_profile, U_heatloss=0e1)

solarr, _, _ = reactor.simulate()
```

```python
from pyreactorkit.display import plot_species, plot_temperature, plot_reactions

plot_species(solarr)
plot_temperature(solarr)

plt.figure(figsize=(6, 3))
plot_reactions(solarr, ax=plt.gca())
```
